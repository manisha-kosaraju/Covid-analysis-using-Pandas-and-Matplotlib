```python
#3.What are the top 5 states with the highest number of deaths?
result=df.groupby(by=['State/UnionTerritory']).sum()
```

```python
sorted_result=result.sort_values(by=["Deaths"],ascending=False)
y=sorted_result.nlargest(5,'Deaths')
y
```

```python
#piechart representation for the top 5 states with the highest number of deaths
y.plot(kind="pie", y="Deaths", startangle=90, figsize=(8, 8))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(y.index, y['Deaths'],color='skyblue')
plt.xlabel('State/UnionTerritory')
plt.ylabel('Total No of Deaths')
plt.title('Top 5 states with the highest number of deaths')
#rotate x axis labels for better readability
plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
plt.show()
```

```python
# Reset index to convert the index values into a regular column
y_reset_index = y.reset_index()
y_reset_index.plot(kind='barh', x='State/UnionTerritory', y='Deaths', color='red')

# Add labels and title
plt.xlabel('State/UnionTerritory', color='blue')
plt.ylabel('Total No of Deaths',color='green')
plt.title('Top 5 states with the highest number of deaths', color='purple')
#need not explicitly define the below two functions ie legend() and show().
#plt.legend()
#plt.show()
```

```python
y_reset_index.plot(kind='line',x='State/UnionTerritory', y='Deaths', color='purple', ls=':')
plt.xlabel('State/UnionTerritory', color='blue')
plt.ylabel('Total No of Deaths',color='green')
plt.title('Top 5 states with the highest number of deaths', color='purple')
```

```python
#4.In which month, there were hightest no of confirmed cases?
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df
```

```python
#fill the na values to 0.
#df['Month'].fillna(0, inplace=True)
df.fillna({'Month': 0}, inplace=True)
```

```python
#converting the float datatype to int datatype
df['Month'] = df['Month'].astype(int)
#Here coverting the numerical month to the month name(abbreviations)
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df
```

```python
'''grouping by the name of the month,summing up the confirmed column.Also sorting in descending order.Thus obtaining the highest number 
of confirmed cases.'''
month=df.groupby(by=['Month'])['Confirmed'].sum().sort_values(ascending=False)
```

```python
month
```

```python
result =month.head(1)
result
```

```python
print("The month that has got highest no of confirmed cases:",result.index[0])
```
